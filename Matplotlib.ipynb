```python
## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
## plot a scatter plot.
## Scatter plot uses dots to reprensents the two different numerical variable.
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y, c="#24EA39")
plt.title("Scatter plot of X and Y")
plt.xlabel("this is X axis")
plt.ylabel("this is Y axis")
```

```python
## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

## subplots creates a figure and a grid of subplots with a single call, 
# while providing reasonable control over how the individual plots are created.

#For line 1:
x = np.array([0, 1, 2, 3, 4, 5])   
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(1,4,1)
plt.plot(x,y)
#For line 2:
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(1,4,2)
plt.plot(x,y)
#For line 3:
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(1,4,3)
plt.plot(x,y)
#For line 4: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(1,4,4)
plt.plot(x,y)
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

# Bar plots are a type of data visualization used to represent data in the form of rectangular bars. 
# The height of each bar represents the value of a data point, and the width of each bar represents the category of the data.
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit,color="#3924EA")
```

```python
## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

## box plot displays the five-number summary of a set of data. 
## The five-number summary is the minimum, first quartile, median, third quartile, and maximum. 
## In a box plot, we draw a box from the first quartile to the third quartile.
## A vertical line goes through the box at the median.
## Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
```
